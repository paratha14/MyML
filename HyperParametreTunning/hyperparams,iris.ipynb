```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
iris= load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['flower']=iris.target
df['flower']=df.flower.apply(lambda x : iris.target_names[x])
df
```

```python
#using searchgridcv method
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
clf= GridSearchCV(LogisticRegression(solver='liblinear',multi_class='auto'),
                  {'C':[1,5,10]},
                  cv=3,
                  return_train_score=False
                  )
clf.fit(iris.data,iris.target)
```

```python
clf.cv_results_
```

```python
result=pd.DataFrame(clf.cv_results_)
result
```

```python
result[['param_C','mean_test_score']]
```

```python
#now using Randomisedsearchgrid
rsg= RandomizedSearchCV(LogisticRegression(solver='liblinear',multi_class='auto'),{'C':[1,7,11]},cv=5,return_train_score=False,n_iter=2)
rsg.fit(iris.data,iris.target)
```

```python
re= pd.DataFrame(rsg.cv_results_)
re
```

```python
re[['param_C','mean_test_score']]
```

```python
#now for hyperparametre tunning

model_params={


    'forest':{
        'model':RandomForestClassifier(),
        'params': {
            'n_estimators':[1,7,12]
        }
    },

    'logisticregression':{
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{'C':[1,7,12]}
    }
}
```

```python
score= []
for mn,mp in model_params.items():
  clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
  clf.fit(iris.data,iris.target)
  score.append({
      'model':mn,
      'best_params':clf.best_params_,
      'best_score':clf.best_score_
  })

```

```python
df=pd.DataFrame(score)
```
