```python
import matplotlib.pyplot as plt
```

```python
setosa= df[df['target']=='setosa']
versicolor= df[df['target']=='versicolor']
virginica= df[df['target']=='virginica']
```

```python
setosa.head()
```

```python
plt.title('Setosa vs Versicolor')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], color='green', marker='+')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], color='blue', marker='*')
plt.show()
```

```python
plt.title('Setosa vs virginca')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], color='green', marker='+')
plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], color='blue', marker='*')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(df.drop(['target'], axis='columns'), df.target, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.svm import SVC
model= SVC(C=10)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
yp= model.predict(X_test)
```

```python
len(yp)
```

```python
len(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, yp)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
plt.show()
```
