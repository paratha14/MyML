```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits= load_digits()
```

```python
dir(digits)
```

```python
df= pd.DataFrame(digits.data, columns= digits.feature_names)
```

```python
target= digits.target
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X= df
X_scaled= sc.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_scaled, target, test_size= 0.2, random_state=50)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(100)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
pca= PCA(0.95)
X_pca= pca.fit_transform(X)
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test= train_test_split(X_pca, target, test_size= 0.2, random_state=50)
```

```python
model_pca= RandomForestClassifier(500)
model_pca.fit(X_train_pca, y_train)
```

```python
model_pca.score(X_test_pca, y_test)
```

```python
yp_pca = model_pca.predict(X_test_pca)
```

```python
yp= model.predict(X_test)
```

```python
#prediction using randomforest
plt.xlabel('actual')
plt.ylabel('predicted')
plt.plot(y_test, yp_pca, color="green", marker= "*", label= "through pca")
plt.plot(y_test, yp, color= "blue", marker= "^", label= "random forest" )
plt.legend()
plt.show()
```
