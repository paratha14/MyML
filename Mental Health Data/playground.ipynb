```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("train.csv")
train
```

```python
test=pd.read_csv("test.csv")
test
```

```python
#dropping columns that are not valuable for prediction.
train=train.drop(['Name','City',],axis='columns')
test=test.drop(['Name','City'],axis='columns')
```

```python
#creating dummy variables
train_gender=pd.get_dummies(train.Gender)
test_gender=pd.get_dummies(test.Gender)
train_work=pd.get_dummies(train['Working Professional or Student'])
test_work=pd.get_dummies(test['Working Professional or Student'])

#merging them to their respective datasets
train=pd.concat([train,train_gender,train_work],axis='columns')
test=pd.concat([test,test_gender,test_work],axis='columns')

#dropping unrequired columnns
target=train.Depression
train=train.drop(['Gender','Working Professional or Student','Depression'],axis='columns')
test=test.drop(['Gender','Working Professional or Student'],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le_tpro=LabelEncoder()
le_tsleep=LabelEncoder()
le_tdiet=LabelEncoder()
le_tdeg=LabelEncoder()
le_ts=LabelEncoder()
le_tm=LabelEncoder()

#fitiing data into train
train['Profession_n']=le_tpro.fit_transform(train['Profession'])
train['Sleep Duration_n']=le_tsleep.fit_transform(train['Sleep Duration'])
train['Dietary Habits_n']=le_tdiet.fit_transform(train['Dietary Habits'])
train['Degree_n']=le_tdeg.fit_transform(train['Degree'])
train['suicide_n']=le_ts.fit_transform(train['Have you ever had suicidal thoughts ?'])
train['mental_n']=le_tm.fit_transform(train['Family History of Mental Illness'])

#fiiting data into test
le_ttpro=LabelEncoder()
le_ttsleep=LabelEncoder()
le_ttdiet=LabelEncoder()
le_ttdeg=LabelEncoder()
le_tts=LabelEncoder()
le_ttm=LabelEncoder()

test['Profession_n']=le_ttpro.fit_transform(test['Profession'])
test['Sleep Duration_n']=le_ttsleep.fit_transform(test['Sleep Duration'])
test['Dietary Habits_n']=le_ttdiet.fit_transform(test['Dietary Habits'])
test['Degree_n']=le_ttdeg.fit_transform(test['Degree'])
test['suicide_n']=le_tts.fit_transform(test['Have you ever had suicidal thoughts ?'])
test['mental_n']=le_ttm.fit_transform(test['Family History of Mental Illness'])
```

```python
#dropping old columns
test=test.drop(['Profession','Sleep Duration','Dietary Habits','Degree','Have you ever had suicidal thoughts ?','Family History of Mental Illness'],axis='columns')
train=train.drop(['Profession','Sleep Duration','Dietary Habits','Degree','Have you ever had suicidal thoughts ?','Family History of Mental Illness'],axis='columns')
```

```python
#settin NA values to 0.0 (in train)

train['Academic Pressure']=train['Academic Pressure'].fillna(0)
train['Work Pressure']=train['Work Pressure'].fillna(0)
train['CGPA']=train['CGPA'].fillna(0)
train['Study Satisfaction']=train['Study Satisfaction'].fillna(0)
train['Job Satisfaction']=train['Job Satisfaction'].fillna(0)
train['Financial Stress']=train['Financial Stress'].fillna(0)
train.isna().any()


```

```python
#setting NA values to 0.0 (in test)

test['Academic Pressure']=test['Academic Pressure'].fillna(0)
test['Work Pressure']=test['Work Pressure'].fillna(0)
test['CGPA']=test['CGPA'].fillna(0)
test['Study Satisfaction']=test['Study Satisfaction'].fillna(0)
test['Job Satisfaction']=test['Job Satisfaction'].fillna(0)
test['Financial Stress']=test['Financial Stress'].fillna(0)
test.isna().any()
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(solver='liblinear')
reg.fit(train,target)
```

```python
#storing all predicted values in a variable y_pre
y_pre=reg.predict(test)
```

```python
#creating a new dataframe and then creating the required columns
final=pd.DataFrame()
final['id']=test['id']
final['Depression']=y_pre
```
