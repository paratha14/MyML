```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
training_df= pd.read_csv("/content/drive/MyDrive/train.csv")
training_df.head(10)
```

```python
training_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()

training_df["TrackDurationMs"]= scaler.fit_transform(training_df[["TrackDurationMs"]])
training_df["AudioLoudness"]= scaler.fit_transform(training_df[["AudioLoudness"]])
```

```python
training_df.describe()
```

```python
training_df.isna().any()
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
X_train= training_df.drop(["id", "BeatsPerMinute"], axis=1)
y_train= training_df["BeatsPerMinute"]
```

```python
model.fit(X_train, y_train)
```

```python
testing_df= pd.read_csv("/content/drive/MyDrive/test.csv")
testing_df.head(10)
```

```python
testing_df["TrackDurationMs"]= scaler.fit_transform(testing_df[["TrackDurationMs"]])
testing_df["AudioLoudness"]= scaler.fit_transform(testing_df[["AudioLoudness"]])
```

```python
testing_df.describe()
```

```python
testing_df.isna().any()
```

```python
testing_df.describe()
X_test= testing_df.drop(["id"], axis=1)
```

```python
yp= model.predict(X_test)
```

```python
result=pd.DataFrame({"id": testing_df["id"], "BeatsPerMinute": yp})
```
